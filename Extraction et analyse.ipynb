```python
#créer une fonction pour avoir un diactionnaire {mot:fois}
def dico_mot(words):
    word_counts = {}
    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

    return word_counts
```

```python
#obtenir les mots
def findWord(listMot):
    return re.findall("[a-zA-ZÀ-ÿ][a-zA-ZÀ-ÿ-]*", listMot)
```
